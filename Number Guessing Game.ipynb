```python
import random
logo = """
  / _ \_   _  ___  ___ ___  /__   \ |__   ___    /\ \ \_   _ _ __ ___ | |__   ___ _ __ 
 / /_\/ | | |/ _ \/ __/ __|   / /\/ '_ \ / _ \  /  \/ / | | | '_ ` _ \| '_ \ / _ \ '__|
/ /_\\| |_| |  __/\__ \__ \  / /  | | | |  __/ / /\  /| |_| | | | | | | |_) |  __/ |   
\____/ \__,_|\___||___/___/  \/   |_| |_|\___| \_\ \/  \__,_|_| |_| |_|_.__/ \___|_|  
"""
def easy(n,a):
    print("You Have 10 Attempts To Guess The Number.")
    win=0
    while win==0 and a!=0:
        inp3='y'
        while inp3=='y':
            guess=int(input("Make a Guess: "))
            if guess>100 and guess<1:
                print("Number Out Of Range!! ") 
                inp3='y'
            else:
                inp3='n'
        if guess>n:
            print("Too High")
            print("Guess Again")
            a=a-1
            print(f"You Have {a} attempts remaining.")
        elif guess<n:
            print("Too Low")
            print("Guess Again")
            a=a-1
            print(f"You Have {a} attempts remaining.")
        elif guess==n:
            print(f"You Got It!! The Answer was {n}.")
            win=1
            welcome()
    if a==0:
        print("You Ran Out Of Guesses!! You Lose!!")
        print(f"The Number was {n}")
        welcome()
def hard(n,a):
    print("You Have 5 Attempts To Guess The Number.")
    win=0
    while win==0 and a!=0:
        inp3='y'
        while inp3=='y':
            guess=int(input("Make a Guess: "))
            if guess>100 and guess<1:
                print("Number Out Of Range!! ") 
                inp3='y'
            else:
                inp3='n'
        if guess>n:
            print("Too High")
            print("Guess Again")
            a=a-1
            print(f"You Have {a} attempts remaining.")
        elif guess<n:
            print("Too Low")
            print("Guess Again")
            a=a-1
            print(f"You Have {a} attempts remaining.")
        elif guess==n:
            print(f"You Got It!! The Answer was {n}.")
            win=1
            welcome()
    if a==0:
        print("You Ran Out Of Guesses!! You Lose!!")
        print(f"The Number was {n}")
        welcome()
def welcome():
    print(logo)
    print("Welcome To The Number Guessing Game.")
    print("I'm thinking of a number between 1 and 100")
    num=random.randint(1,100)
    inp2='y'
    while inp2=='y':
        inp=input("Choose A Difficulty 'easy' or 'hard': ")
        inp=inp.lower()
        if inp=='easy':
            attempts=10
            easy(num,attempts)
        elif inp=='hard':
            attempts=5
            hard(num,attempts)
        else:
            inp2=input("Invalid Input!! Do you wish to continue? (Y - Yes | N - No): ")
            inp2=inp2.lower()
welcome()
```
